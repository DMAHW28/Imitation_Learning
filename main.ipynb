```python
from models import EvaluateAllModel
import matplotlib.pyplot as plt
import numpy as np
```

## Cart Pole

```python
model_eval_cartPole = EvaluateAllModel(max_epochs=10, num_round_gail = 65)
```

#### Behavioral Cloning stats

```python
fig, axs = plt.subplots(1, 2, figsize=(30, 10))

axs[0].plot(range(1, len(model_eval_cartPole.bc_reward_in_learning) + 1), model_eval_cartPole.bc_reward_in_learning,label="rewards")
axs[0].set_xlabel("Epoch")
axs[0].set_ylabel("rewards")
axs[0].set_title(f"rewards over Epochs in learning for {model_eval_cartPole.max_transitions_episodes*500} demos Behaviral Cloning")
axs[0].legend()

demo_number = np.array(range(1, len(model_eval_cartPole.bc_reward_in_demo) + 1)) * 500
axs[1].plot(demo_number, model_eval_cartPole.bc_reward_in_demo, label="rewards")
axs[1].set_xlabel("number of demo")
axs[1].set_ylabel("rewards")
axs[1].set_title(f"rewards over number of demo  for {model_eval_cartPole.max_epochs} epochs Behaviral Cloning")
axs[1].legend()
plt.show()
```

#### GAIL stats

```python
fig, axs = plt.subplots(1, 2, figsize=(30, 10))

axs[0].plot(range(1, len(model_eval_cartPole.gail_reward_in_learning) + 1), model_eval_cartPole.gail_reward_in_learning,label="rewards")
axs[0].set_xlabel("round")
axs[0].set_ylabel("rewards")
axs[0].set_title(f"rewards over round in learning Gail for {model_eval_cartPole.max_transitions_episodes*500} demos")
axs[0].legend()

axs[1].plot(range(1, len(model_eval_cartPole.gail_reward_in_learning_2) + 1), model_eval_cartPole.gail_reward_in_learning_2,label="rewards")
axs[1].set_xlabel("round")
axs[1].set_ylabel("rewards")
axs[1].set_title(f"rewards over round in learning Gail for {model_eval_cartPole.max_transitions_episodes*500*5} demos")
axs[1].legend()
plt.show()
```
